```python
import os
def get_audio_paths(directory_path):
    audio_paths = []
    # Iterate over the files in the directory
    for filename in os.listdir(directory_path):
        # Get the complete path of the file
        file_path = os.path.join(directory_path, filename)
        # Check if it's a file and not a directory
        if os.path.isfile(file_path):
            # Append the file path to the list
            audio_paths.append(file_path)
    return audio_paths
```

```python
# we need to distinguish the unique labels in our SER dataset
label_list = train_dataset.unique(output_column)
label_list.sort()  # Let's sort it for determinism
num_labels = len(label_list)
print(f"A classification problem with {num_labels} classes: {label_list}")
```

```python
model_checkpoint = "facebook/wav2vec2-base"
#model_checkpoint = "facebook/w2v-bert-2.0"
# model_checkpoint = "/content/drive/MyDrive/Arricor/Audio_deepFake_output_dir/wav2vec2-xlsr-greek-speech-emotion-recognition-celebvox-randomized-Exp2/checkpoint-6395"
batch_size = 10
```

```python
from transformers import AutoConfig, Wav2Vec2Processor
config = AutoConfig.from_pretrained(
    model_checkpoint,
    num_labels=num_labels,
    label2id={label: i for i, label in enumerate(label_list)},
    id2label={i: label for i, label in enumerate(label_list)},
    # finetuning_task="wav2vec2_clf",
)
```

```python
from transformers import AutoModelForAudioClassification, TrainingArguments, Trainer

model = AutoModelForAudioClassification.from_pretrained(
    model_checkpoint,
    num_labels=num_labels,
    label2id=config.label2id,
    id2label=config.id2label,
)
```

```python
from transformers import AutoFeatureExtractor

feature_extractor = AutoFeatureExtractor.from_pretrained(model_checkpoint, padding= True, return_attention_mask = True, return_tensors = "pt")
feature_extractor
```

```python
from dataclasses import dataclass
from typing import Dict, List, Optional, Union
import torch

import transformers
from transformers import Wav2Vec2Processor


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


@dataclass
class CustomDataCollator:

    def __call__(self, inp):
        print(inp)
        features = [  {'input_values' :feature_extractor( [this_datapoint['input_values'][0]['array']])} for this_datapoint in inp]
        batch = feature_extractor.pad(
                                      features,
                                      return_tensors="pt",
                                      padding=True
                                    )
        labels = [config.label2id[this_datapoint['labels']] for this_datapoint in inp ]
        batch['labels'] = torch.tensor( list( labels ) , dtype = torch.long)
        return batch
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

```python
model_name = model_checkpoint.split("/")[-1]
batch_size = 16

training_output_dir = "/content/drive/MyDrive/Arricor/Audio_deepFake_output_dir/wav2vec2-xlsr-greek-speech-emotion-recognition"
args = TrainingArguments(
    output_dir=training_output_dir,
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=batch_size,
    gradient_accumulation_steps=4,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=5,
    warmup_ratio=0.1,
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    push_to_hub=False,
)
```

```python
import numpy as np
from datasets import load_dataset, load_metric

metric = load_metric("accuracy")
def compute_metrics(eval_pred):
    """Computes accuracy on a batch of predictions"""
    predictions = np.argmax(eval_pred.predictions, axis=1)
    return metric.compute(predictions=predictions, references=eval_pred.label_ids)
```

```python
trainer = Trainer(
    model,
    args,
    data_collator=collator,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=feature_extractor,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```
